```python

from bs4 import BeautifulSoup as bs
from splinter import Browser
import pandas as pd
import os
import time
import requests
```

```python
executable_path = {'executable_path': 'chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://nces.ed.gov/ccd/tables/ACGR_RE_and_characteristics_2015-16.asp"
browser.visit(url)
```

```python
html = browser.html
```

```python
tables = pd.read_html(url)

df = tables[1]
df.head()
```

```python
new_header = ['State', 'Total', "American Indian/Alaska Native", 'NaN', 'Asian/Pacific Islander', 'Asian', 'Hawaiian Native/Pacific Islander',
             'NaN2','Hispanic', 'NaN3','Black', 'NaN4', 'White', 'NaN5', 'Two or more races', 'NaN6', 'Economically disadvantaged', 'NaN7', 
             'Limited English Proficiency', 'NaN8', 'Students with disabilities']
df.columns = new_header
df.head()
```

```python
df_clean = df.drop(columns = ['NaN', 'NaN2','NaN3','NaN4','NaN5','NaN6','NaN7','NaN8'])
df_clean.head()
```

```python
df_replace = df_clean.replace({"—": "0", ">=90%":"90", "<>": "0", ">=50%":"50", ">=80%":"80", ">=95%": "95", "<50%": "50"})
df_replace
```

```python
clean_df = df_replace.drop(df_replace.index[1])
clean_df
```

```python
clean_df.to_csv('grad_rate_by_race.csv')
```
