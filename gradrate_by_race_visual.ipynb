```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
grad_rate_csv = 'grad_rate_by_race.csv'
grad_rate = pd.read_csv(grad_rate_csv)
grad_rate.head()
```

```python
x = grad_rate["State"]

plt.figure(figsize = (30,10))
plt.plot(x, grad_rate["Total"], '-r', label= "Total", linewidth=5 )
plt.plot(x, grad_rate["American Indian/Alaska Native"], '-b', label = "American Indian/Alaska Native")
plt.plot(x, grad_rate["Asian/Pacific Islander"], '-g', label ="Asian/Pacific Islander")
plt.plot(x, grad_rate["Hispanic"], '-m', label= "Hispanic")
plt.plot(x, grad_rate["Black"], '-y', label= "Black")
plt.plot(x, grad_rate["White"], '-c', label= "White")

plt.xlabel("State")
plt.ylabel("Graduation Rate")
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

```python
x = grad_rate["State"]

plt.figure(figsize = (30,10))
plt.plot(x, grad_rate["Total"], '-r', label= "Total", linewidth=5 )
plt.plot(x, grad_rate["Economically disadvantaged"], '-b', label = "Economically disadvantaged")
plt.plot(x, grad_rate["Limited English Proficiency"], '-g', label ="Limited English Proficiency")
plt.plot(x, grad_rate["Students with disabilities"], '-m', label= "Students with disabilities")

plt.xlabel("State")
plt.ylabel("Graduation Rate")
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```
